### Markdown Files in Jupyter

If you wish to contribute to the content of this book, you need to modify the
source file (md file, not ipynb file) on GitHub.
Using the notedown plugin we
can modify notebooks in the md format directly in Jupyter.


First, install the notedown plugin, run the Jupyter Notebook, and load the plugin:

```
pip install d2l-notedown  # You may need to uninstall the original notedown.
jupyter notebook --NotebookApp.contents_manager_class='notedown.NotedownContentsManager'
```

You may also turn on the notedown plugin by default whenever you run the Jupyter Notebook.
First, generate a Jupyter Notebook configuration file (if it has already been generated, you can skip this step).

```
jupyter notebook --generate-config
```

Then, add the following line to the end of the Jupyter Notebook configuration file (for Linux or macOS, usually in the path `~/.jupyter/jupyter_notebook_config.py`):

```
c.NotebookApp.contents_manager_class = 'notedown.NotedownContentsManager'
```

After that, you only need to run the `jupyter notebook` command to turn on the notedown plugin by default.

## How to interpret the Regression Table

The regression table provides key statistics to understand the relationship between the independent and dependent variables. Here's a breakdown of the important components:

*   **Dep. Variable**: This is your dependent variable, the variable you are trying to predict.
*   **Model**: The type of regression model used (e.g., OLS - Ordinary Least Squares).
*   **R-squared**: This value represents the proportion of the variance in the dependent variable that is predictable from the independent variable(s). A higher R-squared indicates a better fit of the model to the data.
*   **Adj. R-squared**: This is a modified version of R-squared that has been adjusted for the number of predictors in the model. It's useful when comparing models with different numbers of predictors.
*   **F-statistic**: This is a statistical test that determines whether the overall regression model is statistically significant. A high F-statistic and a low p-value (Prob (F-statistic)) indicate that the model is statistically significant.
*   **coef**: These are the regression coefficients. The coefficient for the independent variable indicates the change in the dependent variable for a one-unit change in the independent variable, holding other variables constant. The coefficient for the constant (const) is the expected value of the dependent variable when all independent variables are zero.
*   **std err**: This is the standard error of the coefficient, which measures the variability of the coefficient estimate.
*   **t**: This is the t-statistic, which is the coefficient divided by its standard error. It is used to test the hypothesis that the coefficient is equal to zero.
*   **P>|t|**: This is the p-value for the t-statistic. A low p-value (typically less than 0.05) indicates that the coefficient is statistically significant, meaning that the independent variable is likely related to the dependent variable.
*   **[0.025 0.975]**: This is the 95% confidence interval for the coefficient. It represents the range within which the true coefficient is likely to fall 95% of the time.

By examining these values, you can assess the strength and significance of the relationship between your variables and the overall fit of your regression model.

SST (Total Sum of Squares): This represents the total variation in the dependent variable (y) around its mean. It measures how much the data points vary from the average value of the dependent variable.
SSR (Sum of Squares Regression): This represents the variation in the dependent variable that is explained by the regression model (i.e., the variation that is attributed to the relationship between the dependent and independent variables). It measures how much the predicted values from the regression line vary from the mean of the dependent variable.

The relationship between these is typically expressed as:

SST = SSR + SSE

Understanding these values helps in evaluating the performance of a regression model, particularly in calculating the R-squared value, which is SSR/SST and represents the proportion of the variance in the dependent variable that is predictable from the independent variable(s).



## Ordinary Least Squares (OLS)

Ordinary Least Squares (OLS) is a common method for estimating the coefficients of linear regression models. The goal of OLS is to find the line that best fits the data by minimizing the sum of the squared differences between the observed values of the dependent variable and the values predicted by the linear model. These differences are called residuals.

Here's a breakdown of the key ideas behind OLS:

1.  **The Linear Model:** OLS assumes a linear relationship between the dependent variable (y) and one or more independent variables (x). This relationship is represented by a linear equation:

    $y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + ... + \beta_n x_n + \epsilon$

    Where:
    *   $y$ is the dependent variable.
    *   $\beta_0$ is the y-intercept (the value of y when all independent variables are zero).
    *   $\beta_i$ are the coefficients for each independent variable $x_i$ (representing the change in y for a one-unit change in $x_i$, holding other variables constant).
    *   $\epsilon$ is the error term (representing the unexplained variation in y).

2.  **Minimizing the Sum of Squared Residuals:** OLS aims to find the values of the coefficients ($\beta_0, \beta_1, ..., \beta_n$) that minimize the sum of the squared residuals. The residual for each data point is the difference between the actual observed value ($y_i$) and the value predicted by the model ($\hat{y}_i$):

    $e_i = y_i - \hat{y}_i$

    The sum of squared residuals (SSR) is calculated as:

    $SSR = \sum_{i=1}^{N} e_i^2 = \sum_{i=1}^{N} (y_i - \hat{y}_i)^2$

    OLS finds the coefficients that minimize this SSR.

3.  **Assumptions of OLS:** For the results of OLS to be reliable and for statistical inference (like hypothesis testing) to be valid, several assumptions should ideally be met:

    *   **Linearity:** The relationship between the dependent and independent variables is linear.
    *   **Independence:** The observations are independent of each other.
    *   **Homoscedasticity:** The variance of the residuals is constant across all levels of the independent variables.
    *   **Normality:** The residuals are normally distributed.
    *   **No Multicollinearity:** The independent variables are not highly correlated with each other.

OLS is a fundamental technique in regression analysis and provides a straightforward way to estimate the relationships between variables.

## Normality Assumption in OLS

One of the key assumptions of Ordinary Least Squares (OLS) regression is that the residuals (the differences between the observed and predicted values) are normally distributed. This assumption is important for several reasons, particularly for performing statistical inference, such as constructing confidence intervals and conducting hypothesis tests on the regression coefficients.

Here's what the normality assumption means and why it's important:

*   **What it means:** The distribution of the residuals should resemble a normal (or Gaussian) distribution, which is a symmetrical, bell-shaped curve. This means that the residuals are centered around zero, and most of the residuals are close to zero, with fewer residuals further away.

*   **Why it's important:**
    *   **Statistical Inference:** Normality of residuals is crucial for the validity of hypothesis tests (like the t-test for coefficients and the F-test for the overall model) and the construction of confidence intervals. If the residuals are not normally distributed, the p-values and confidence intervals may not be accurate, leading to incorrect conclusions about the significance of the predictors.
    *   **Efficiency of Estimates:** While OLS provides unbiased estimates of the coefficients even with non-normal residuals (under certain conditions), the estimates are most efficient (i.e., have the lowest variance) when the residuals are normally distributed.

*   **How to check for normality:** You can assess the normality of residuals using various methods:
    *   **Histograms:** Plotting a histogram of the residuals can visually show if the distribution is roughly bell-shaped.
    *   **Q-Q Plots:** A Q-Q (Quantile-Quantile) plot compares the quantiles of the residuals to the quantiles of a theoretical normal distribution. If the residuals are normally distributed, the points should fall approximately along a straight line.
    *   **Statistical Tests:** Formal statistical tests like the Shapiro-Wilk test or the Kolmogorov-Smirnov test can provide a more objective assessment of normality. However, these tests can be sensitive to sample size.

*   **What to do if residuals are not normal:** If the residuals deviate significantly from normality, you might consider:
    *   **Data Transformations:** Transforming the dependent variable or some of the independent variables can sometimes help to achieve normality.
    *   **Non-parametric methods:** If transformations don't work, you might consider using non-parametric regression methods that do not rely on the assumption of normality.

While the normality assumption is important for valid statistical inference, OLS estimates of the coefficients are still unbiased and consistent even if the residuals are not normal, provided that other assumptions (like linearity and independence) are met. However, for reliable hypothesis testing and confidence intervals, checking and addressing non-normality is recommended.

## Understanding the R-squared Measure

R-squared (R) is a statistical measure that represents the proportion of the variance in the dependent variable that is predictable from the independent variable(s) in a regression model. In simpler terms, it tells you how well your regression model fits the data.

Here's a breakdown of the key aspects of R-squared:

*   **Definition:** R-squared is formally defined as the ratio of the explained variance (variance accounted for by the regression model) to the total variance in the dependent variable. It is often expressed as a percentage.

*   **Formula:** R-squared can be calculated using the following formula:

    $R^2 = \frac{SSR}{SST}$

    Where:
    *   SSR is the Sum of Squares Regression (explained variance).
    *   SST is the Total Sum of Squares (total variance).

    Alternatively, it can be calculated as:

    $R^2 = 1 - \frac{SSE}{SST}$

    Where:
    *   SSE is the Sum of Squares Error (unexplained variance).

*   **Interpretation:**
    *   R-squared values range from 0 to 1 (or 0% to 100%).
    *   An R-squared of 0 indicates that the model explains none of the variability in the dependent variable around its mean. In this case, the model provides no better fit than simply using the mean of the dependent variable.
    *   An R-squared of 1 (or 100%) indicates that the model explains all of the variability in the dependent variable. This means that the model perfectly predicts the dependent variable, which is rare in practice, especially with real-world data.
    *   An R-squared value between 0 and 1 indicates the proportion of the dependent variable's variance that is predictable from the independent variable(s). For example, an R-squared of 0.60 means that 60% of the variation in the dependent variable can be explained by the model's predictors.

*   **Goodness of Fit:** R-squared is often used as a measure of the goodness of fit of a regression model. A higher R-squared value generally indicates a better fit, as more of the variation in the dependent variable is explained by the model.

*   **Limitations:** While R-squared is a useful metric, it has some limitations:
    *   **Does not indicate causality:** A high R-squared does not necessarily mean that the independent variables cause the changes in the dependent variable. Correlation does not imply causation.
    *   **Does not indicate model correctness:** A high R-squared does not guarantee that the model is correctly specified or that the assumptions of regression have been met.
    *   **Increases with more predictors:** Adding more independent variables to a model will always increase the R-squared, even if the new variables are not statistically significant. This is why adjusted R-squared is often used when comparing models with different numbers of predictors.
    *   **Low R-squared does not always mean a bad model:** In some fields, particularly in social sciences, even a low R-squared can be considered meaningful if the predictors are statistically significant and the model is theoretically sound.

*   **Adjusted R-squared:** Adjusted R-squared is a modified version of R-squared that accounts for the number of predictors in the model. It penalizes the addition of unnecessary predictors, making it a more reliable measure when comparing models with different numbers of independent variables. Adjusted R-squared will always be less than or equal to R-squared.

In summary, R-squared is a valuable tool for understanding how much of the variation in your dependent variable is explained by your regression model. However, it should be interpreted in conjunction with other diagnostic statistics and domain knowledge.
